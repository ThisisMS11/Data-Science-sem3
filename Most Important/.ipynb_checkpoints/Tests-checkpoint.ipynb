## T  Tests 
## Independent Samples T-Test


```python
nA=int(input("Enter the number of students in class A :-"))
meanA=float(input("Enter the average marks obtained by students of class A :- "))
sdA=float(input("Enter the standard deviation :- "))
```

```python
nB=int(input("Enter the number of students in class B :-"))
meanB=float(input("Enter the average marks obtained by students of class B :- "))
sdB=float(input("Enter the standard deviation :- "))
```

```python
alpha=float(input("Enter the value for alpha : "))
```

```python
dfTotal=nA+nB-2
```

```python
df1=nA-1
df2=nB-1

ss1= pow(sdA,2)*df1
ss2= pow(sdB,2)*df2

print(ss1)
print(ss2)

sp2=(ss1 + ss2)/(df1 + df2)

print('sp2 = ', sp2)
```

```python
import math
Tstat=(meanA-meanB)/math.sqrt((sp2/nA) + (sp2/nB))
print('Tstat = ',Tstat)
```

```python
Ttable_value= 2.0167
```

```python
if(Tstat < -Ttable_value):
    print("The Hypothsis must be rejected")
elif(Tstat > Ttable_value):
    print("The Hypothesis must be rejected")
else:
    print("The Hypothesis must not be rejected")
```

## Dependent Sample T-Test![image.png](attachment:image.png)

```python
before=[9,10,7,5,7,5,9,6,8,7]
after=[7,6,5,4,4,6,7,5,5,7]

n=len(before)

dif=[]
for i in range(n):
    dif.append(before[i]-after[i])
dif
```

```python
# Now find the standard deviation using the normal formula 
sumsquare=0
sumdif=0
for i in dif:
    sumsquare+=pow(i,2)
    sumdif+=i

sd=(sumsquare-(pow(sumdif,2)/n))/(n-1)
sd=math.sqrt(sd)


mean=sumdif/n

tstat=mean*math.sqrt(n)/sd
```

```python
tTable=2.2622
```

```python
if(tstat< -tTable):
    print("Hypothesis must be rejeceted")
elif(tstat > tTable):
    print("Hypothesis must be rejected")
else:
    print("Hypothesis must not be rejected")
```

# Chi-Square Test for Goodness of Fit

```python
observed=[121,288,91]
expected=[100,150,250]
n= len(observed)
df=n-1
alpha=0.05
tTable_value=5.99
```

```python
chieStat=0
for i in range(n):
    hold=pow((observed[i]-expected[i]),2)
    chieStat+=(hold/expected[i])
chieStat
```

```python
if(chieStat> tTable_value):
    print("Reject the Null hypothesis")
```

# One way Anova

```python
a=[8,10,12,8,7]
b=[12,11,9,14,4]
c=[18,12,16,6,8]
d=[13,9,12,16,15]

N=len(a)+len(b)+len(c)+len(d)

sum_a=0;
squared_sum_a=0;

sum_b=0;
squared_sum_b=0

sum_c=0
squared_sum_c=0

sum_d=0
squared_sum_d=0

Grand_Total=0

for i in a:
    sum_a+=i
    squared_sum_a+=pow(i,2)
Grand_Total+=sum_a

mean_a=sum_a/len(a)

for i in b:
    sum_b+=i
    squared_sum_b+=pow(i,2)
Grand_Total+=sum_b

mean_b=sum_b/len(b)

for i in c:
    sum_c+=i
    squared_sum_c+=pow(i,2)
Grand_Total+=sum_c

mean_c=sum_c/len(c)


for i in d:
    sum_d+=i
    squared_sum_d+=pow(i,2)
    
Grand_Total+=sum_d

mean_d=sum_d/len(d)

correction_factor=pow(Grand_Total,2)/N

print(sum_a)
print(squared_sum_a)
print(sum_b)
print(squared_sum_b)
print(sum_c)
print(squared_sum_c)
print(sum_d)
print(squared_sum_d)
```

```python
# print(Grand_Total)
correction_factor=pow(Grand_Total,2)/N
# print(correction_factor)


#sum of squares betweeen sample (SSC)
SSC=((pow(sum_a,2)/len(a))+(pow(sum_b,2)/len(b))+(pow(sum_c,2)/len(c))+ (pow(sum_d,2)/len(d))) - (pow(Grand_Total,2)/N)
# df = no_of_columns(C)- 1
df1=3

MSC=SSC/df1

print(SSC)
```

```python
# Sum of Square of Total
SST=squared_sum_a + squared_sum_b + squared_sum_c + squared_sum_d - correction_factor
# print(SST) 258
```

```python
# sum of square within sample (SSE)
SSE = SST- SSC
print(SSE)

df2=N-4
MSE=SSE/df2 # 13
print(MSE)
```

```python
F_ratio= MSC/MSE
print(F_ratio)
```
