```python
import pandas as pd
import numpy as np
```

```python
file=pd.read_csv('test.csv')
```

```python
filehead=file.head()
```

```python
filehead
```

```python
for i in filehead:
    holdcol=filehead[i]
    isnull=holdcol.isnull()
    
    mean=0
    nullcounter=0
    
    for nullvalue,value in zip(isnull,holdcol):
        if(nullvalue==True):
            nullcounter+=1
        else:
            mean+=value
            
    mean/=(len(filehead)-nullcounter)
    print(mean)
    
    filehead[i].fillna(value=mean, inplace=True)
#     ignore the warning
```
